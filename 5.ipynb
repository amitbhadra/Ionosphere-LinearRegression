```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ionosphere.data')
tf = df[300:351]
df = df[0:300]
tf
```

```python
#### ==========Part 1===========#

import cvxpy as cp

'''Training'''
df_np = df.to_numpy()
for i in range(0,np.size(df_np,0)):
    if(df_np[i][34]=='b'):
        df_np[i][34] = -1
    elif(df_np[i][34]=='g'):
        df_np[i][34] = 1
n = 34

Y = [df_np[:,34]]
X = df_np[:,0:34]
beta = cp.Variable(1)
w = cp.Variable((34,1))

cost = cp.sum_squares(X@w + beta - Y)
res = cp.Problem(cp.Minimize(cost)).solve()


'''Testing'''
tf_np = tf.to_numpy()
loss = 0

for i in range(np.size(tf_np,0)):
        if(tf_np[i][34]=='b'):
            tf_np[i][34] = -1
        elif(tf_np[i][34]=='g'):
            tf_np[i][34] = 1

for i in range((np.size(tf_np, 0))):
    param = tf_np[i][:-1]
    expected = tf_np[i][-1]

    res = param @ w.value + beta.value
    if((res[0]>=0 and expected>=0) or (res[0]<0 and expected<0)): pass
    else: loss+=1
    #loss += abs(round(res[0],0)-expected)

print(1-loss/51)
```
